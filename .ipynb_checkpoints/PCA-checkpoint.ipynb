```python
import numpy as np
import pandas as pd
import yfinance as yf

import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

from hmmlearn.hmm import GaussianHMM
```

```python
# read demeaned sp500 data
data = pd.read_csv('SP500_demeaned_log_return.csv', index_col = False)
```

```python
# generate covariance matrix
data_var = data.cov()
```

```python
pca = PCA()
pca.fit(data_var)

# Get the eigenstocks (principal components)
eigenstocks = pca.components_.T

# Get the explained variance ratios
explained_variance_ratios = pca.explained_variance_ratio_

# Select the first 10 eigenstocks and explained variance ratios
eigenstocks_first_10 = eigenstocks[:, :10]
explained_variance_first_10 = explained_variance_ratios[:10]

# Generate the x-axis values (time points)
time_points = np.arange(eigenstocks_first_10.shape[0])  # Use shape[0] for the number of time points

# Plot the eigenstocks and their explained variance ratios
plt.figure(figsize=(10, 6))  # Adjust the figure size as per your preference

for i in range(10):
    plt.subplot(5, 2, i+1)
    plt.plot(time_points, eigenstocks_first_10[:, i], label='Eigenstock {}'.format(i+1))
    plt.axhline(0, color='black', linestyle='--', linewidth=0.5)  # Add a horizontal line at y=0
    plt.xlabel('Time')
    plt.ylabel('Value')
    plt.title('Eigenstock {} (Explained Variance: {:.2f}%)'.format(i+1, explained_variance_first_10[i] * 100))
    plt.legend()

plt.tight_layout() 
plt.show()
```

```python
# Calculate the cumulative explained variance
cumulative_variance = np.cumsum(explained_variance_ratios)

# Plot the cumulative explained variance
plt.figure(figsize=(24, 18)) 
plt.plot(range(1, len(cumulative_variance) + 1), cumulative_variance, marker='o')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('Cumulative Explained Variance Plot')
plt.grid(True)
plt.show()
```

```python
# choose number of components so that cum explained var is at least 95%
n = np.argmax(cumulative_variance >= 0.99) + 1
print(n)

# re-run PCA 
pca = PCA(n_components = n)
pca.fit(data_var)
transformed_cov = pca.transform(data_var)
```

```python
date = data['Unnamed: 0']
data = data.drop(data.columns[0], axis=1)
data = data.dropna()
tsla_returns = data['TSLA']

# Project the original data onto the first 10 principal components
projected_data = np.dot(data.values, eigenstocks_first_10)

# Split the data into a training set and a test set
X_train, X_test, y_train, y_test = train_test_split(projected_data, tsla_returns, test_size=0.2, random_state=42)

# Train the regression model
reg = LinearRegression()
reg.fit(X_train, y_train)

# Use the model to predict TSLA's returns
y_pred = reg.predict(X_test)
```

```python
# Get the last known price
tsla_p = pd.read_csv('TSLA_data.csv', index_col = False)['Close']

split_index = int(len(data) * 0.8)

last_known_price = tsla_p.iloc[split_index - 1]
```

```python
dates =date.tail(82)

# Convert the test and predicted data to a comparable scale
y_test_prices = np.cumprod(1 + y_test) * last_known_price
y_pred_prices = np.cumprod(1 + y_pred) * last_known_price

# Create a new figure
fig, ax = plt.subplots(figsize=(14, 8))

# Convert dates to datetime if not already
dates = pd.to_datetime(dates)

# Plot actual and predicted prices
ax.plot(dates, y_test_prices.values, label='Actual prices')
ax.plot(dates, y_pred_prices, label='Predicted prices')

# Setting x-axis major locator as per the frequency of dates you have (weekly in this example)
ax.xaxis.set_major_locator(mdates.WeekdayLocator(interval=1))  

# Setting x-axis major formatter to display dates in a readable format
ax.xaxis.set_major_formatter(mdates.DateFormatter('%b %d')) 

# Adding title and labels
ax.set_title('TSLA: Actual vs Predicted Prices')
ax.set_xlabel('Time')
ax.set_ylabel('Price')
ax.legend()

# Show the plot
plt.show()
```
