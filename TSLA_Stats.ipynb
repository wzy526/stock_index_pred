```python
import numpy as np
import pandas as pd
import yfinance as yf
```

```python
import matplotlib.pyplot as plt
from pandas.plotting import table
```

```python
#read tsla price data 
tsla = pd.read_csv('TSLA_data.csv')[['Date','Close']]
tsla['Return'] = tsla['Close'].pct_change()

#use the stdv of return for standardizatoin purpose
tsla['Return'].describe()
```

```python
#create a subplot without frame
plot = plt.subplot(111, frame_on=False)

t_plot = tsla['Return'].describe()

#remove axis
plot.xaxis.set_visible(False) 
plot.yaxis.set_visible(False) 

#create the table plot and position it in the upper left corner
table(plot,t_plot,loc='upper right')

#save the plot as a png file
plt.savefig('tsla_plot.png')
```

```python
#compare with sp500 data
sp500 = pd.read_csv('SP500_data.csv')[['Date','Close']]
sp500['Return'] = sp500['Close'].pct_change()

sp500['Return'].describe()
```

```python
#create a subplot without frame
plot = plt.subplot(111, frame_on=False)

s_plot = sp500['Return'].describe()

plot.xaxis.set_visible(False) 
plot.yaxis.set_visible(False) 

table(plot, s_plot,loc='upper right')

plt.savefig('sp500_plot.png')
```
